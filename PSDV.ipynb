```python
''' Mutables are data type which we can edit directly,whereas immutables are the data types which we can't edit directly but can edit after 
several steps.'''
l=[1,2,3,4,5]  # list is mutable
t=(1,2,3,4,5)  # tuple is immutable
l.append(6)
print(l)
'''t.append(6) # 'tuple' object has no attribute'append'
print(t)'''
# we cannot append an element directly in immutables(tuple), but we can append by converting into mutable.
x=list(t)
x.append(6)
t=tuple(x)
print(t)
```

```python
import matplotlib.pyplot as plt
def f(n):
    l=[0,1]
    x = 0
    y = 1
    c = 0
    if n == 0:
        return [0]
    elif n == 1:
        return [1]
    else:
        for i in range(2,n):
            s = x + y
            l.append(s)
            c += 1
            if c % 2 == 1:
                x = s
            else:
                y = s
        print(l)
        return l
n=int(input())
plt.figure(figsize=(10,5))
plt.plot(range(n),f(n),marker='o',color='blue',linestyle='-')
plt.xlabel('n',fontsize=12)
plt.ylabel('Fibonacci Number(f(n))',fontsize=12)
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

```python
def f(n):
    m=min(n)
    M=max(n)
    l=[]
    if m==M:
        for i in range(len(n)):
            l.append(0.5)
    else:
        for i in n:
            l.append((i-m)/(M-m))
    return(l)
l= list(map(int,input().split()))
print(f(l))
```
